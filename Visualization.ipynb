If we want to visualize a data set, which questions do we ask?



```python
import pandas
iris = pandas.read_csv('../data/iris.csv')
iris.head()
```

### Which species has the largest petal lengths?

```python
iris.groupby('Species').mean()
```

### How do petal lengths relate to petal widths?

Important questions:

- What are the variables involved?   petal length, petal width.
- What are the variable types? **numerical** and **numercial**.  Therefore, a good plot is the scatter plot.

Visualization in Python.

- Not perfect. No one solution that solves everything.
- Good packages: seaborn, plotly, pandas (basic plots).
- Many good (easy to use API) packages are built on top of matplotlib.



```python
from matplotlib import pyplot
iris.plot.scatter(x='PetalWidth', y='PetalLength')
pyplot.show()
```

Same question but for different species.

We have another variable: Species.  This is a **categorical** variable.



```python
setosa, versicolor, virginica = iris[ iris.Species=='setosa' ], iris[ iris.Species=='versicolor'], iris[ iris.Species=='virginica']
```

```python
ax = setosa.plot.scatter(x='PetalWidth', y='PetalLength', label='Setosa', color='Red')
ax = versicolor.plot.scatter(x='PetalWidth', y='PetalLength', label='Versicolor', color='Blue', ax=ax)
ax = virginica.plot.scatter(x='PetalWidth', y='PetalLength', label='Virginica', color='Green', ax=ax)
```

```python
import seaborn
seaborn.relplot(x='PetalWidth', y='PetalLength', hue='Species', data=iris)
```

```python
seaborn.relplot(x='PetalWidth', y='PetalLength', hue='Species', size=5*iris.SepalLength, data=iris)

```
